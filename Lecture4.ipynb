<h4> Higher dimensional spaces

```python
import numpy as np
import matplotlib.pyplot as plt
```

- 2D Gauss

```python
means = np.zeros(2)
sigma_cov = np.identity(2)
N = 100
data = np.random.multivariate_normal(means, sigma_cov, 100) 
distances = []
for i in range(N):
    for j in range(i,N):
        distances.append(np.linalg.norm(data[i]-data[j]))
        
plt.hist(distances)
plt.show()
```

<h4> Eigenvalues and eigenvectors

```python
from numpy import linalg as LA

A = np.array([[1,2],
              [5,6]])
lam, v = LA.eig(A) 

print(lam) #eigenvalues
print()
print(v) #eigenvectors (columns)
```

```python
print(np.dot(A,v[:,0]))
print()
print(lam[0]*v[:,0])
```

<h4> PageRank alogorithm

```python
import numpy as np
from numpy import linalg as LA

A = np.array([[0,0,0,0,0.5,0],[0.25,0,0,0,0,0],[0,0.5,0,0,0,0],[0.25,0.5,0,0,0.5,0],[0.25,0,1,1,0,1],[0.25,0,0,0,0,0]])
lam, v = LA.eig(A) 

print(lam)

import matplotlib.pyplot as plt

#polar coordinates

def polar(z):
    a= z.real #x coordinate
    b= z.imag #y coordinate
    r = np.sqrt(a**2+b**2) #radius
    theta = np.arctan2(b,a) #angle
    return r,theta 


for x in lam:
    plt.polar([0,polar(x)[1]],[0,polar(x)[0]],marker='o')

print()
print(np.abs(lam)) #we take the highest eigenvalue
print()
print(np.abs(v[:,0])) #... and its corresponding eigenvector
```
